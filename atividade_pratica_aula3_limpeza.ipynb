## 5. Batalha 2 — O enigma dos valores ausentes

A aula reforça que valores ausentes não devem ser tratados automaticamente; a decisão precisa ser justificada. fileciteturn2file0

### Tarefa
1. Mapeie os valores ausentes por coluna
2. Identifique quais colunas críticas têm nulos
3. Defina uma estratégia para cada caso:
   - remover linhas?
   - preencher?
   - manter?
4. Justifique cada escolha

### Dica
Evite decisões automáticas sem análise de contexto.


```python
# Investigue os nulos e aplique aqui sua estratégia de tratamento

#Manter a coluna de data, pois é o mais seguro a se fazer em relação às analises posteriores e a integridade do dataset


df_vendas['canal'].replace({'Loja fisica':'Loja Física','ONLINE':'Online','MarketPlace': 'Marketplace'}, inplace=True)
df_vendas['categoria'].replace({'telefonia':'Telefonia','Perifericos':'Periféricos'}, inplace=True)
df_vendas.dropna(subset=['canal'],inplace=True)
df_vendas.dropna(subset=['categoria'],inplace=True)
# Substituindo valores  nas tabelas canal e categoria e apagando valores nulos. Isso faz com que as análises e agrupamentos sejam feitas corretamente

df_vendas[df_vendas['receita'].isnull() | df_vendas['lucro'].isnull()] #Verificando que dados são esses
# Apagar dados com receita = 0(toda venda deve ser concluída), Pedidos lucrativos sem receita(receita depende de lucro)
receita_nula = df_vendas[df_vendas['receita'].isnull()]
df_vendas = df_vendas.drop(receita_nula.index) # Apagando as linhas em que a receita é nula

lucro_nulo = df_vendas[df_vendas['lucro'].isnull()]
df_vendas = df_vendas.drop(lucro_nulo.index) # Apagando as linhas em que o lucro é nulo

df_vendas[df_vendas['observacao'].isnull()] # Observações nulas não farão diferença, se eliminadas
df_vendas.dropna(subset=['observacao'],inplace=True)
df_vendas.isna().sum()
```
